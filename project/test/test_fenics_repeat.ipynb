```python
"""Testing how many times it seems reasonable to reuse a 
particular FEniCS solution - choosing a different random 
location each time - as we don't want too much overlap
between the different locations chosen.
"""
```

```python
import numpy as np

import ipympl
import matplotlib.pyplot as plt
import matplotlib.patches as patches

%run base_test.ipynb

from base import grid
```

```python
left_intervals_t = np.floor((grid.num_intervals.t - 1) / 2)
left_intervals_x = np.floor((grid.num_intervals.x - 1) / 2)

left_amount_t = left_intervals_t * grid.coarse_grid_sep.t
left_amount_x = left_intervals_x * grid.coarse_grid_sep.x

amount_t = (grid.num_intervals.t + 1) * grid.coarse_grid_sep.t
amount_x = (grid.num_intervals.x + 1) * grid.coarse_grid_sep.x
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
ax.set_xlim(0, 20)
ax.set_ylim(0, 10)

for i in range(1000):
    t_point = np.random.uniform(grid.num_intervals.t * grid.coarse_grid_sep.t, 
                                10 - grid.num_intervals.t * grid.coarse_grid_sep.t)
    x_point = np.random.uniform(0 + grid.num_intervals.x * grid.coarse_grid_sep.x, 
                                20 - grid.num_intervals.x * grid.coarse_grid_sep.x)
                                
    # Coarse grids
    #rect = patches.Rectangle((x_point - left_amount_x, t_point - left_amount_t), amount_x, amount_t, fill=False)
    # Fine grids
    rect = patches.Rectangle((x_point, t_point), grid.coarse_grid_sep.x, grid.coarse_grid_sep.t)
    ax.add_patch(rect)
```
