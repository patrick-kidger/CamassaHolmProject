```python
%run base_test.ipynb

import json
import multiprocessing as mp
import numpy as np
import os

import base.fenics as fc
```

```python
def create_solution(i):
    point, sol = fc.FenicsSolution.gen()
    base_str = '../fenics_data/{}/'.format(i)
    os.mkdir(base_str)
    np.save(base_str + 'uvals', sol.uvals)
    # Don't really need to save all of this, it's true.
    # Some of it can be reconstructed simply from the
    # other information saved.
    # But this is particular straightforward.
    np.save(base_str + 'tvals', sol.tvals)
    np.save(base_str + 'xvals', sol.xvals)
    with open(base_str + 'other_data', 'w') as f:
        f.write(json.dumps({'fineness_t': sol.fineness_t, 
                            'fineness_x': sol.fineness_x,
                            't': sol.t,
                            'T': sol.T,
                            'a': sol.a,
                            'b': sol.b,
                            'initial_condition': sol.initial_condition}))
```

```python
pool = mp.Pool(processes=None)
pool.map_async(create_solution, range(600))
```
